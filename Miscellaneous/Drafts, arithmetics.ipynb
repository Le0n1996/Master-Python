```python
for a in range(1, 6):
    for b in range(-5, 6):
        for c in range(-5, round(b**2/(4*a))):
            for d in range(1, 6):
                for e in range(-5, 6):
                    for f in range(-5, round(e**2/(4*d))):
                        n = (b**2-4*a*c)**0.5
                        if int(n) == float(n):
                            n2 = (e**2-4*d*f-2*d/a*(b+n))**0.5
                            n3 = (e**2-4*d*f-2*d/a*(b-n))**0.5
                            if int(n2) == float(n2):
                                if int(n3) == float(n3):
                                    print(a, b, c, d, e, f)
```
